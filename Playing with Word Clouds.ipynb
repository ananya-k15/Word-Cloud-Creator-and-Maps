<h1> Mickey Mouse Word Cloud </h1> 

```python
# opening the file and read it into a variable 
mickey_text = open('mickey_mouse.txt', 'r').read()
    
print ('File downloaded and saved!')
```

Next, let's use the stopwords that we imported from `word_cloud`.

```python
# instantiate a word cloud object
stopwords = set(STOPWORDS)
mickey_wc = WordCloud(
    background_color='white',
    max_words=4000,
    stopwords=stopwords
)

# generate the word cloud
mickey_wc.generate(mickey_text)
```

Visualizing the word cloud -

```python
# display the word cloud
plt.imshow(mickey_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Let's resize the cloud so that we can see the less frequent words a little better.

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(mickey_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Excellent! This looks really interesting! Another cool thing you can implement with the `word_cloud` package is superimposing the words onto a mask of any shape. Let's use a mask mickey mouse.

```python
mickey_mask = np.array(Image.open('mickey.jpg'))
    
print('Image downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
mickey_wc = WordCloud(
    background_color='white',
    max_words=4000,
    stopwords=stopwords
)

# generate the word cloud
mickey_wc.generate(mickey_text)
```

Let's take a look at how the mask looks like.

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(mickey_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def colors(word=None, font_size=None,
                     position=None, orientation=None,
                     font_path=None, random_state=None):
    colors = [[4, 77, 82],
              [25, 74, 85],
              [82, 43, 84],
              [158, 48, 79]]
    rand = random_state.randint(0, len(colors) - 1)
    return "hsl({}, {}%, {}%)".format(colors[rand][0], colors[rand][1], colors[rand][2])
```

```python
# instantiate a word cloud object
wc_mickey = WordCloud(background_color='white', max_words=2000, mask=mickey_mask, stopwords=stopwords, color_func=colors)

# generate the word cloud
wc_mickey.generate(mickey_text)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(10) # set width
fig.set_figheight(10) # set height

plt.imshow(wc_mickey, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Really impressive!

```python
from wordcloud import ImageColorGenerator
```

```python
mask_colors = ImageColorGenerator(mickey_mask)
wc_mickey = WordCloud(
               mask=mickey_mask, background_color="white",
               max_words=2500, max_font_size=256,
               random_state=42, width=mickey_mask.shape[1],
               height=mickey_mask.shape[0], color_func=mask_colors)
fig = plt.figure()
fig.set_figwidth(7) # set width
fig.set_figheight(9)
wc_mickey.generate(mickey_text)
plt.imshow(wc_mickey, interpolation="bilinear")
plt.axis('off')
plt.show()
```
